# Local Minimum

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
def f_local_min(x, y):
    return (x**2 + y**2 + 1)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f_local_min(X, Y)
```

Function: $f(x,y)= x^2 + y^2 + 1$

$f_x(x,y) = 2x \qquad f_y(x,y) = 2y$ 

$2x=0 \qquad 2y=0 \qquad$ Critical Point at (0,0)

$f_{xx}(x,y)=2 \qquad f_{yy}(x,y) = 2 \qquad f_{xy}(x,y) = 0$

$Let D = D(0,0) = f_{xx}(0,0) * f_{yy}(0,0) - f_{xy}^2(0,0)  = (2)(2) + 0^2 = 4$

Since $D > 0$ and $f_{xx}(0,0) = 2$, a local minimum exist

```
plt.contour(X, Y, Z, 30, cmap='BuGn')
plt.colorbar();
plt.title('2D Contour Plot $f(x,y)=x^2 + y^2 + 1$')
plt.plot(0,0,"or");
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 20, cmap='binary')
ax.scatter3D(0,0,1, c='red')
ax.set_xlabel('x')
ax.set_ylabel('y')
# ax.set_zlabel('z');
ax.set_title('3D Contour Plot $f(x,y)=x^2 + y^2 + 1$');
```

```
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Surface Plot $f(x,y)=x^2 + y^2 + 1$');
```

# Local Maximum

```
def f_local_max(x, y):
    return (-x**2 - y**2 + 35)

X1, Y1 = np.meshgrid(x, y)
Z1 = f_local_max(X1, Y1)
```

Function: $f(x,y)= -x^2 - y^2 + 35$

$f_x(x,y) = -2x \qquad f_y(x,y) = -2y$ 

$-2x=0 \qquad -2y=0 \qquad$ Critical Point at (0,0)

$f_{xx}(x,y)= -2 \qquad f_{yy}(x,y) = -2 \qquad f_{xy}(x,y) = 0$

$Let D = D(0,0) = f_{xx}(0,0) * f_{yy}(0,0) - f_{xy}^2(0,0)  = (-2)(-2) + 0^2 = 4$

Since $D > 0$ and $f_{xx}(0,0) = -2$ local maximum exist

```
plt.contour(X1, Y1, Z1, 30, cmap='BuGn')
plt.colorbar();
plt.plot(0,0,"or")
plt.title('2D Contour Plot $f(x,y)= -x^2 - y^2 + 35$');
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, Y1, Z1, 20, cmap='binary')
ax.scatter3D(0,0,35, c='red')
ax.set_xlabel('x')
ax.set_ylabel('y')
# ax.set_zlabel('z')
ax.set_title('3D Contour Plot $f(x,y)=-x^2 - y^2 + 35$');
```

```
ax1 = plt.axes(projection='3d')
ax1.plot_surface(X1, Y1, Z1, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax1.set_title('Surface Plot $f(x,y)= -x^2 - y^2 + 35$');
```

# Neither Local Minimum or Local Maximum

```
def f_saddle(x, y):
    return (x**2 - y**2 + 2)
  
X2, Y2 = np.meshgrid(x, y)
Z2 = f_saddle(X2, Y2)
```

# f(x,y) = x^2 - y^2 + 2
$f_x(x,y) = 2x \qquad f_y(x,y) = -2y$ 

$2x=0 \qquad -2y=0 \qquad$ Critical Point at (0,0)

$f_{xx}(x,y)= 2 \qquad f_{yy}(x,y) = -2 \qquad f_{xy}(x,y) = 0$

$Let D = D(0,0) = f_{xx}(0,0) * f_{yy}(0,0) - f_{xy}^2(0,0)  = (2)(-2) + 0^2 = -4$

Since $D < 0$ saddle exist at critical point

```
plt.contour(X2, Y2, Z2, 30, cmap='BuGn')
plt.colorbar();
plt.plot(0,0,"or");
plt.title('Contour Plot $f(x)= x^2 - y^2 + 2$');
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X2, Y2, Z2, 20, cmap='binary')
ax.scatter3D(0,0,2, c='red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('3D Contour Plot $f(x)= x^2 - y^2 + 2$');
```

```
ax2 = plt.axes(projection='3d')
ax2.plot_surface(X2, Y2, Z2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax2.set_title('Surface Plot  $f(x)= x^2 - y^2 + 2$');
```

# Quiver - Local Minimum

```
# f(x,y) = x^2 + y^2 + 1
U = 2*X
V = 2*Y
fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)
ax.quiverkey(q, X=0.3, Y=1.1, U=10, label='Quiver key, length = 10', labelpos='E');

plt.show();
```

# Quiver - Local Maximum

```
# f(x,y) = -x^2 -y^2 + 35
U1 = -2 * X1
V1 = -2 * Y1
fig, ax = plt.subplots()
q = ax.quiver(X1, Y1, U1, V1)
ax.quiverkey(q, X=0.3, Y=1.1, U=10, label='Quiver key, length = 10', labelpos='E');

plt.show()
```
