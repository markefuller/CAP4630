# Displaying First 10 MNIST digits


```
import matplotlib.pyplot as plt
```

```
f, axarr = plt.subplots(1,10,figsize=(10, 10))
for i in range(10):
   axarr[i].imshow(train_images[i]);
```

# Group Digits by Label and Display 10 of Each


```
import numpy as np

digits = []
for i in range (10):
  digits.append(train_images[train_labels == i])
  
# Display the first 10 images for each grouping
fig, axes = plt.subplots(nrows=10, ncols=10, figsize=(10, 10))
for j in range(10):
    for i in range(10):
        axes[j,i].imshow(digits[j][i])
```

# Display Average for Each Digit

```
from PIL import Image
for i in range(10):
    arr = np.array(np.mean(digits[i], axis=(0)), dtype=np.uint8)
    out = Image.fromarray(arr)
    plt.figure(figsize=(1, 1))
    plt.imshow(out)
    plt.show()
```
